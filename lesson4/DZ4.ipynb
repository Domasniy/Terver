### Задание 1
Случайная непрерывная величина A имеет равномерное распределение на промежутке (200, 800].
Найдите ее среднее значение и дисперсию.

$$M(X) = \frac{a + b}{2} = \frac{200+800}{2}=\frac{1000}{2}=500$$
$$D(X) = \frac{(b - a)^2}{12} = \frac{(800-200)^2}{12}=\frac{(600)^2}{12}=30000$$

$$f(x) = \frac{1}{4 \sqrt{2 \pi}} e ^ {- \frac{(x + 2)^2}{32}}$$

- M(X)
- D(X)

### Задание 4
Рост взрослого населения города X имеет нормальное распределение.

Причем, средний рост равен 174 см, а среднее квадратичное отклонение равно 8 см.

Какова вероятность того, что случайным образом выбранный взрослый человек имеет рост:

- больше 182 см
- больше 190 см
- от 166 см до 190 см
- от 166 см до 182 см
- от 158 см до 190 см
- не выше 150 см или не ниже 190 см
- не выше 150 см или не ниже 198 см
- ниже 166 см.

$$\mu = 174 $$
$$\sigma = 8 $$

Будем находить Z-score по формуле: $$ Z = \frac{x - \mu}{\sigma}$$

И дальше пользоватся Z таблицей

a) больше 182 см

$$Z = \frac{182-174}{8}= 1$$
$$p = 1 - 0.8413 = 0.1587(15.87\%)$$

б) больше 190 см

$$Z = \frac{190-174}{8}= 2$$
$$p = 1 - 0.9772 = 0.0228(2.28\%)$$

в) от 166 см до 190 см

Найдем Z-score для 190
$$Z = \frac{190-174}{8}= 2$$
Найдем Z-score для 166
$$Z = \frac{166-174}{8}= -1$$
$$p = 0.9772 - 0.1587 = 0.8185(81.85\%)$$

г) от 166 см до 182 см

Найдем Z-score для 182
$$Z = \frac{182-174}{8}= 1$$
Найдем Z-score для 166
$$Z = \frac{166-174}{8}= -1$$
$$p = 0.8413 - 0.1587 = 0,6826(68.26\%)$$

д) от 158 см до 190 см

Найдем Z-score для 190
$$Z = \frac{190-174}{8}= 2$$
Найдем Z-score для 158
$$Z = \frac{158-174}{8}= -2$$
$$p = 0.9772 - 0.0228 = 0,9544(95.44\%)$$

e) не выше 150 см или не ниже 190 см

Найдем Z-score для 190
$$Z = \frac{190-174}{8}= 2$$
Найдем Z-score для 150
$$Z = \frac{150-174}{8}= -3$$
$$p = (1-0.9772) + 0.0013 = 0,0241(2.41\%)$$


ё) не выше 150 см или не ниже 198 см

Найдем Z-score для 198
$$Z = \frac{198-174}{8}= 3$$
Найдем Z-score для 150
$$Z = \frac{150-174}{8}= -3$$
$$p = (1-0.9987) + 0.0013 = 0,0026(0.26\%)$$

ж) ниже 166 см.

Найдем Z-score для 166
$$Z = \frac{166-174}{8}= -1$$

$$p = 0,1587(15.87\%)$$
